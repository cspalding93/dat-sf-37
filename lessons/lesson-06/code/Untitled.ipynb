```python
import numpy as np

def rg_bin():
    secret_prob = 1 / float(0b0100)
    
    if secret_prob < np.random.rand():
        return 'r'
    else:
        return 'g'
```

```python
sample = [rg_bin() for _ in range(10)]

sample
```

```python
acc = [1,1,0,1]

def boot():
    return np.mean(np.random.choice(acc,size=len(acc),replace=True))
    
accMeans = [boot() for _ in range(100)]
```

```python
%matplotlib inline
import seaborn as sns
```

```python
sns.distplot(accMeans)
```
