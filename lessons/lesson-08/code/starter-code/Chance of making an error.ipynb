```python
import numpy as np

def ten_coins():
    num_heads = 0
    for i in range(10):
        if np.random.rand() >= .5:
            num_heads += 1
    return num_heads
```

```python
samples = [ten_coins() for _ in range(10000)]
```

```python
%matplotlib inline
import seaborn as sns

sns.distplot(samples)
```

```python
def bootstrap_sample():
    bootstrap_samples = np.random.choice(samples,len(samples),replace=True)
    return np.mean(bootstrap_samples)
```

```python
bootstrap_samples = [bootstrap_sample() for _ in range(10000)]
```

```python
sns.distplot(bootstrap_samples)
```
