```python
from numpy.random import rand
```

```python
import numpy as np
```

```python
def delivery_time():
    def source_del():
        rn = rand()
        return 2**int(rn*4)
    source_del_time = []
    for i in range(4):
        source_del_time.append(source_del())
    
    return np.max(source_del_time)
```

```python
%matplotlib inline
```

```python
delivery_time()
```

```python
samples = [delivery_time() for i in range(10000)]
```

```python
np.mean(samples)
```

```python
import seaborn as sns
```

```python
sns.distplot(samples)
```

```python
np.std(samples)
```

```python
def almost_normal():
    return sum(np.random.uniform(size=200))

dist = np.array([almost_normal() for _ in range(10000)])

sns.distplot(dist,bins = 50)
```
