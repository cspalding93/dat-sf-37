```python
# Using word2vec similarity scores
peace_tweets = []
for tweet in tweets[:200]:
    parsed = nlp_toolkit(tweet)
    filter_words = ['war','iran']
    war_words = model.most_similar(positive=['war'])
    iran_words = model.most_similar(positive=['iran'])
    for word in war_words:
        filter_words.append(word[0])
        
    for word in iran_words:
        filter_words.append(word[0])
        
    for word in parsed:
        if word.text in filter_words:
            continue
            
    peace_tweets.append(tweet)

for tweet in peace_tweets[:10]:
    print tweet
```
